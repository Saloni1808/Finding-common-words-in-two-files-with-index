```python
word_list=[]
with open("positive.txt","r") as pos:
    for line in pos:
        line=line.lower().replace("\n","")
        word_list.append(line)
#print(word_list)
```

```python
import pandas as pd
data=pd.read_csv("res-har.csv")
csv_data=list(data.body)
```

```python
#print(type(list(csv_data)))
for i in word_list:
    index_list=[]
    for j in csv_data:
        if i in j.lower():
            index_list.append(csv_data.index(j))
    if index_list:
        #pass
        print(i,"found in lines",index_list) #check +2 for csv
```
